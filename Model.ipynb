**Musig generation using AI**

```python
import random
from music21 import stream,note,midi
def generate_melody(prompt,length=16):
  words_to_notes={
      "happy":["C5","E5","G5","A5"],
      "sad":["C4","E4","G4","A4"],
      "mysterious":["D4","F4","A4","B4"],
      "exciting":["C6","D6","E6","G6"],
  }
  notes=words_to_notes.get(prompt.lower(),["C4","D4","E4","G4"])
  melody=stream.Stream()
  for _ in range(length):
    pitch=random.choice(notes)
    duration=random.choice([0.5,1.0,1.5])
    melody.append(note.Note(pitch=pitch,quarterLength=duration))
  return melody

def save_melody(melody,filename="generated.midi"):
  mf=midi.translate.music21ObjectToMidiFile(melody)
  mf.open(filename,"wb")
  mf.write()
  mf.close()
  print(f"Music saved as {filename}")

prompt=input("Enter a prompt: ")
melody=generate_melody(prompt)
save_melody(melody)
print("AI generated music created!")
```

**Image and Video generatioin using AI**

```python
import torch
import cv2
import os
import numpy as np
from diffusers import StableDiffusionPipeline

print(torch.cuda.is_available())  # Should print True

#load the diffusion model
model_name="runwayml/stable-diffusion-v1-5"
pipe=StableDiffusionPipeline.from_pretrained(model_name,torch_dtype=torch.float16)
pipe=pipe.to("cuda")
output_folder="generated_frames"
os.makedirs(output_folder,exist_ok=True)
```

```python
def generate_images(prompt,num_images=10):
  image_paths=[]
  for i in range(num_images):
    print(f"Generating frame{i+1}/{num_images}...")
    image=pipe(prompt).images[0]
    image_path=os.path.join(output_folder,f"frame_{i:03d}.png")
    image.save(image_path)
    image_paths.append(image_path)
  return image_paths
```

```python

from moviepy.editor import ImageSequenceClip
def create_video(image_paths,output_video="generated_video.mp4",fps=40):
  frame=[cv2.imread(img) for img in image_paths]
  video_clip=ImageSequenceClip(image_paths,fps=fps)
  video_clip.write_videofile(output_video,codec="libx264")
  print(f"Smooth Ai generated video saved as {output_video}")

prompt=input("Enter a prompt: ")
image_paths=generate_images(prompt,num_images=15)
create_video(image_paths,fps=20)
print("AI generated video created")
```
